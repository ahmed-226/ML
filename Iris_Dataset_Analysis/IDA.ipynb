# Iris Dataset Info

 The Iris dataset is a popular dataset in machine learning and statistics. It is often used as a beginner's dataset
 to explore and understand various classification algorithms. The dataset contains measurements of four features of
 three different species of Iris flowers: setosa, versicolor, and virginica.

**The four features measured in the dataset are:**
 1. Sepal length (in centimeters)
 2. Sepal width (in centimeters)
 3. Petal length (in centimeters)
 4. Petal width (in centimeters)

 The goal of using the Iris dataset is to predict the species of an Iris flower based on its measurements. This is
 typically framed as a supervised learning problem, where the input features are the measurements and the target
 variable is the species.

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import os
warnings.filterwarnings("ignore")
```

### Loading the dataset

```python
from sklearn.datasets import load_iris

iris = load_iris()

df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target

df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df['target']=le.fit_transform(df['target'])
df.head()
```

### Modeling

```python
from sklearn.model_selection import train_test_split

X=df.drop(columns=['target'])
Y=df['target']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
# logistic regression

from sklearn.linear_model import LogisticRegression
LCR = LogisticRegression()
LCR.fit(x_train, y_train)
```

```python
print('Accuracy',LCR.score(x_test, y_test)*100)
```

```python
# decision tree classifier
from sklearn.tree import DecisionTreeClassifier
DTC = DecisionTreeClassifier()
DTC.fit(x_train, y_train)
```

```python
print('Accuracy',DTC.score(x_test, y_test)*100)
```
